# Case Study: Fitting peaks in an Infrared (IR) Spectra

In the lesson and exercises you saw an example of an [infrared spectrum for ethanol](https://webbook.nist.gov/cgi/cbook.cgi?ID=C64175&Units=SI&Type=IR-SPEC&Index=2#IR-SPEC). In this case study we will continue to work with the spectrum, but will examine a different region. Let's take a look:

```python
%matplotlib inline
import pandas as pd
from matplotlib import pylab as plt

df = pd.read_csv('data/ethanol_IR.csv')
x_all = df['wavenumber [cm^-1]'].values
y_all = df['absorbance'].values

fig, ax = plt.subplots()
ax.plot(x_all,y_all)
ax.set_xlabel('wavenumber [cm^-1]')
ax.set_ylabel('absorbance')
```

In this case study you will work with the region of the spectrum from ~800 to ~1325 cm$^{-1}$:

```python
x_peak = x_all[75:185]
y_peak = y_all[75:185]

fig, ax = plt.subplots()
ax.plot(x_peak,y_peak, '-b', marker='.')
ax.set_xlabel('wavenumber [cm^-1]')
ax.set_ylabel('absorbance')
max(x_peak)
```
